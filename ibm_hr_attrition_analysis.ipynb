```python
# Advanced analysis
# 1.Attrition vs years at company
df["tenure_group"] = pd.cut(df["years_at_company"],
                          bins=[0,2,5,10,20,40],
                          labels=['0-2','2-5','5-10','10-20','20+'])
attrition_rate = (df["attrition"] == "Yes").groupby(df["tenure_group"]).mean() * 100
attrition_rate.round(2)
```

```python
# Trend analysis
import matplotlib.pyplot as plt
attrition_trend = (df["attrition"] == "Yes").groupby(df["years_at_company"]).mean() * 100
attrition_trend.plot(kind='line', marker='o')
plt.xlabel("Years at Company")      # X-axis label
plt.ylabel("Attrition Rate (%)")    # Y-axis label
plt.title("Attrition vs Years at Company")

plt.show()
```

```python
# 2. Attrition vs job satisfaction

attrition_rate = (df["attrition"] == "Yes").groupby(df["job_satisfaction"]).mean() * 100
attrition_rate.plot(kind='bar')
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition vs Job Satisfaction")
plt.show()
```

```python
# 3. Attrition rate
df["attrition"].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title("Attrition Percentage")
plt.ylabel("")   # removes default label
plt.show()
```

```python
# 4.Boxplot (Salary vs Attrition)
import seaborn as sns
sns.boxplot(x ="attrition",y="monthly_income",data = df)
```
